```python
with open('MPE1004_0000000000.json','r') as f:
    metadata = json.loads(f.read(),encoding='UTF-8') 
```

```python
del metadata['results']
```

```python
pd.json_normalize(metadata)
```

```python
def flatten_json(y):
    out = {}

    def flatten(x, name=''):
        if isinstance(x, dict):
            for a in x:
                flatten(x[a], name + a + '_')
        elif isinstance(x, list):
            i = 0
            for a in x:
                flatten(a, name + str(i) + '_')
                i += 1
        else:
            out[str(name[:-1])] = str(x)

    flatten(y)
    return out
```

```python
metadata_flat = flatten_json(metadata)
```

```python
results_flat = [flatten_json(result) for result in data['results']]
```

```python
def merge(list, dict2): 
    new_list =[]
    i = 0
    for a in list:
        res = {**a, **dict2}
        new_list.append(res)
        i += 1
    return new_list
```

```python
merge_flat = merge(results_flat,metadata_flat)
```

```python
df = pd.DataFrame(merge_flat)
```

```python
df.to_csv('MPE/2020/08/28/MPE1004.csv',index=False)
```
